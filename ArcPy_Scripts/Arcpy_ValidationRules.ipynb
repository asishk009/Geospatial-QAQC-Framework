# Small Length in Lines

```python
import arcpy
import os

folder_path = r"E:\STUDENT\AUCE\GDBs\2 Towns"

arcpy.env.workspace = folder_path
gdbs = arcpy.ListWorkspaces(workspace_type="FileGDB")

header = f"{'CITY_NAME':<20} | {'Short All_lines (<=1m)':<25}"
print(header)
print("-" * len(header))

for gdb in gdbs:
    filename = os.path.basename(gdb)
    city_name = filename.split('_')[0].upper()
    
    try:
        all_lines_path = os.path.join(gdb, "Base_ULU", "All_lines")

        def count_short_lines(fc, length_limit=1):
            if not arcpy.Exists(fc):
                return "N/A"
            
            count = 0
            with arcpy.da.SearchCursor(fc, ["SHAPE@LENGTH"]) as cursor:
                for row in cursor:
                    if row[0] is not None and row[0] <= length_limit:
                        count += 1
            return count

        count_all = count_short_lines(all_lines_path)

        print(f"{city_name:<20} | {count_all:<25}")

    except Exception as e:
        print(f"{city_name:<20} | Error: {e}")

print("-" * len(header))
print("Analysis Complete.")
```

# Dangles in All_lines

```python
import arcpy
import os

arcpy.env.overwriteOutput = True
gdb = r"E:\STUDENT\AUCE\GDBs\2 Towns\SOUNAMURA_Feature Extraction.gdb"

configs = {
    "Base_ULU": {
        "name": "Base_ULU_Topology",
        "rules": [
            ("All_lines", "Must Not Self-Overlap (Line)"),
            ("All_lines", "Must Not Have Dangles (Line)"),
            ("Buildings", "Must Not Overlap (Area)"),
            ("All_polygons", "Must Not Have Gaps (Area)"),
            ("All_polygons", "Must Not Overlap (Area)")
        ]
    },
    "Boundaries": {
        "name": "Boundary_Topology",
        "rules": [
            ("City_AOI", "Must Not Have Gaps (Area)")
        ]
    }
}

for fd_name, config in configs.items():
    fd_path = os.path.join(gdb, fd_name)
    topo_path = os.path.join(fd_path, config["name"])
    
    try:
        print(f"Processing: {fd_name}")
        
        arcpy.management.CreateTopology(fd_path, config["name"])
        
        unique_fcs = list(set([r[0] for r in config["rules"]]))
        
        for fc_name in unique_fcs:
            fc_path = os.path.join(fd_path, fc_name)
            arcpy.management.AddFeatureClassToTopology(topo_path, fc_path, 1, 1)
        
        for fc_name, rule_type in config["rules"]:
            fc_path = os.path.join(fd_path, fc_name)
            print(f"Adding rule: {rule_type} to {fc_name}")
            arcpy.management.AddRuleToTopology(topo_path, rule_type, fc_path)

        print(f"Validating: {config['name']}")
        arcpy.management.ValidateTopology(topo_path)
        
    except Exception as e:
        print(f"Error in {fd_name}: {e}")
```

```python
fd_path = os.path.join(gdb, "Base_ULU")
topo_path = os.path.join(fd_path, configs["Base_ULU"]["name"])
all_lines = os.path.join(fd_path, "All_lines")
final_selection = os.path.join(fd_path, "Dangles_Line_Errors")

temp_point_errs = os.path.join(fd_path, "temp_e_point")
temp_buffer = r"memory\temp_buffer"
temp_join = r"memory\temp_join"

try:
    print("Exporting topology errors...")
    arcpy.management.ExportTopologyErrors(topo_path, fd_path, "temp_e")

    if arcpy.Exists(temp_point_errs):
        print("Buffering points (0.3m)...")
        arcpy.analysis.Buffer(temp_point_errs, temp_buffer, "0.3 Meters")

        print("Performing Spatial Join...")
        arcpy.analysis.SpatialJoin(
            target_features=temp_buffer,
            join_features=all_lines,
            out_feature_class=temp_join,
            join_operation="JOIN_ONE_TO_ONE",
            match_option="INTERSECT"
        )

        print("Filtering Join_Count > 1...")
        lyr = arcpy.management.MakeFeatureLayer(temp_join, "selection_lyr", "Join_Count > 1")
        
        error_count = int(arcpy.management.GetCount(lyr)[0])
        
        if error_count > 0:
            arcpy.management.CopyFeatures(lyr, final_selection)
            print(f"Success: {error_count} Dangles-Line errors saved to {final_selection}")
        else:
            print("No Dangles-Line errors (Join_Count > 1) found.")

        arcpy.management.Delete("memory")
        for s in ["_point", "_line", "_poly"]:
            fc = os.path.join(fd_path, f"temp_e{s}")
            if arcpy.Exists(fc):
                arcpy.management.Delete(fc)
    else:
        print("No point errors available to process.")

except Exception as e:
    print(f"Error: {e}")
```

```python
import arcpy
import os

folder_path = r"E:\STUDENT\AUCE\GDBs\2 Towns"
arcpy.env.workspace = folder_path
gdbs = arcpy.ListWorkspaces(workspace_type="FileGDB")

# Header
print(f"{'City':<15} | {'Count (<= 5sqm)':<15} | {'Type of Feature'}")
print("-" * 65)

for gdb in gdbs:
    filename = os.path.basename(gdb)
    city_name = filename.split('_')[0].upper()
    
    try:
        # Paths based on your data structure
        poly_path = os.path.join(gdb, "Base_ULU", "All_polygons")
        bldg_path = os.path.join(gdb, "Base_ULU", "Buildings")
        
        city_findings = []

        # 1. Process All_polygons (Dynamically find categories)
        if arcpy.Exists(poly_path):
            # Try to find a descriptive field like 'Sub_Class' or 'Category'
            fields = [f.name for f in arcpy.ListFields(poly_path)]
            label_field = next((f for f in fields if f.lower() in ['class', 'category', 'type', 'landuse']), None)
            
            # Group by the label field
            stats = {}
            search_fields = ["SHAPE@AREA", label_field] if label_field else ["SHAPE@AREA"]
            
            with arcpy.da.SearchCursor(poly_path, search_fields) as cursor:
                for row in cursor:
                    area = row[0]
                    # If label field exists, get it; otherwise use "General Polygon"
                    label = row[1] if label_field and row[1] else "General Polygon"
                    
                    if area is not None and area <= 5:
                        stats[label] = stats.get(label, 0) + 1
            
            for label, count in stats.items():
                if count > 0:
                    city_findings.append((count, f"All_polygons - Class {label}"))

        # 2. Process Buildings
        if arcpy.Exists(bldg_path):
            bldg_count = 0
            with arcpy.da.SearchCursor(bldg_path, ["SHAPE@AREA"]) as cursor:
                for row in cursor:
                    if row[0] is not None and row[0] <= 5:
                        bldg_count += 1
            
            if bldg_count > 0:
                city_findings.append((bldg_count, "Buildings"))

        # --- Print Results (Grouped Style) ---
        for i, (count, feat_label) in enumerate(city_findings):
            if i == 0:
                print(f"{city_name:<15} | {count:<15} | {feat_label}")
            else:
                print(f"{'':<15} | {count:<15} | {feat_label}")
        
        if city_findings:
            print("-" * 65)

    except Exception as e:
        print(f"{city_name:<15} | Error: {str(e)[:25]}...")

print("Analysis Complete.")
```

```python
import arcpy
import os

gdb_path = r"E:\STUDENT\AUCE\GDBs\2 Towns\SOUNAMURA_Feature Extraction.gdb"

def scan_gdb_duplicates(target_gdb):
    arcpy.env.workspace = target_gdb
    city_name = os.path.basename(target_gdb).split('_')[0].upper()
    
    print(f"Processing GDB: {city_name}")
    print(f"{'Duplicates':<10} | {'Feature Class'}")
    print("-" * 50)

    walk = arcpy.da.Walk(target_gdb, datatype="FeatureClass")
    
    total_found = 0

    for dirpath, dirnames, filenames in walk:
        for fc_name in filenames:
            fc_path = os.path.join(dirpath, fc_name)
            
            out_table = os.path.join("memory", "temp_identical")
            
            try:
                arcpy.management.FindIdentical(
                    in_dataset=fc_path, 
                    out_dataset=out_table, 
                    fields=["Shape"], 
                    xy_tolerance="0.01 Meters"
                )
                
                total_rows = int(arcpy.management.GetCount(out_table)[0])
                
                if total_rows > 0:
                    sequences = [row[0] for row in arcpy.da.SearchCursor(out_table, "FEAT_SEQ")]
                    duplicate_count = len(sequences) - len(set(sequences))
                    
                    if duplicate_count > 0:
                        print(f"{duplicate_count:<10} | {fc_name}")
                        total_found += duplicate_count

                arcpy.management.Delete(out_table)

            except Exception:
                if arcpy.Exists(out_table):
                    arcpy.management.Delete(out_table)
                continue

    if total_found == 0:
        print("No duplicates found in this GDB.")
    
    print("-" * 50)

if arcpy.Exists(gdb_path):
    scan_gdb_duplicates(gdb_path)
else:
    print("Invalid GDB path provided.")
```

```python
import arcpy
import os
import csv

parent_folder = r"E:\STUDENT\AUCE\GDBs\New folder"
output_csv = r"E:\STUDENT\AUCE\GDBs _Validation_Results\GDBs_Duplicates_6.csv"

arcpy.env.workspace = parent_folder
gdbs = arcpy.ListWorkspaces(workspace_type="FileGDB")

results = {}
all_cities = []
total_gdbs = len(gdbs)

print("Scanning GDBs and identifying duplicates...")
print("-" * 70)

for index, gdb in enumerate(gdbs, start=1):
    gdb_filename = os.path.basename(gdb)
    city_name = gdb_filename.split('_')[0].upper()
    all_cities.append(city_name)
    
    print(f"Processing City: {city_name}")
    
    walk = arcpy.da.Walk(gdb, datatype="FeatureClass")
    
    for dirpath, dirnames, filenames in walk:
        dataset_name = os.path.basename(dirpath) if ".gdb" not in os.path.basename(dirpath).lower() else "None"
        
        if dataset_name not in results:
            results[dataset_name] = {}
            
        for fc_name in filenames:
            if fc_name not in results[dataset_name]:
                results[dataset_name][fc_name] = {}
                
            fc_path = os.path.join(dirpath, fc_name)
            out_table = os.path.join("memory", "temp_dups")
            
            try:
                arcpy.management.FindIdentical(fc_path, out_table, ["Shape"], "0.001 Meters")
                
                total_rows = int(arcpy.management.GetCount(out_table)[0])
                
                if total_rows > 0:
                    with arcpy.da.SearchCursor(out_table, ["FEAT_SEQ"]) as cursor:
                        sequences = [row[0] for row in cursor]
                    
                    duplicate_count = len(sequences) - len(set(sequences))
                    results[dataset_name][fc_name][city_name] = duplicate_count
                else:
                    results[dataset_name][fc_name][city_name] = 0
                    
                arcpy.management.Delete(out_table)
            except:
                results[dataset_name][fc_name][city_name] = 0
                if arcpy.Exists(out_table): arcpy.management.Delete(out_table)

sorted_datasets = sorted(results.keys(), key=lambda x: (x == "None", x))

print("-" * 70)
print(f"Exporting CSV to: {output_csv}")

with open(output_csv, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    header = ["Dataset", "FeatureClass"] + sorted(all_cities)
    writer.writerow(header)
    
    for ds in sorted_datasets:
        for fc in sorted(results[ds].keys()):
            row = [ds, fc]
            for city in sorted(all_cities):
                row.append(results[ds][fc].get(city, 0))
            writer.writerow(row)

print("Batch Process and Export Finished Successfully.")
```

# Bus stop point on Buildings

```python
import arcpy
import os

folder_path = r"E:\STUDENT\AUCE\GDBs\2 Towns"

arcpy.env.workspace = folder_path
gdbs = arcpy.ListWorkspaces(workspace_type="FileGDB")

print(f"{'CITY':<15} | {'Bus Stop Point on Buildings'}")
print("-" * 45)

for gdb in gdbs:
    filename = os.path.basename(gdb)
    city_name = filename.split('_')[0].upper()
    
    try:
        busstop_fc = os.path.join(gdb, "Bus_Stop_Pnt")
        buildings_fc = os.path.join(gdb, "Buildings")
        
        temp_busstop = "temp_busstop"
        arcpy.management.MakeFeatureLayer(busstop_fc, temp_busstop)

        temp_buildings = "temp_buildings"
        arcpy.management.MakeFeatureLayer(buildings_fc, temp_buildings)

        arcpy.management.SelectLayerByLocation(
            in_layer=temp_busstop,
            overlap_type="INTERSECT",
            select_features=temp_buildings,
            selection_type="NEW_SELECTION"
        )

        count = int(arcpy.management.GetCount(temp_busstop)[0])

        print(f"{city_name:<15} | {count}")
        
        arcpy.management.Delete(temp_busstop)
        arcpy.management.Delete(temp_buildings)

    except Exception as e:
        print(f"{city_name:<15} | Error: {e}")
```
